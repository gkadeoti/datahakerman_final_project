```python

def characters_api_call(index):
    url = f"https://anapioficeandfire.com/api/characters/{index}"
    response = requests.get(url)
    result = response.json()
    return result
```

```python
characters_api_call(1)
```

```python
def repeated_api_calls(api_call, start_index, end_index):
    results = []
    for i in range(start_index, end_index + 1):
        result = api_call(i)
        results.append(result)
    return results
```

```python
repeated_api_calls(characters_api_call, 1, 40)
```

```python
characters_40 = repeated_api_calls(characters_api_call, 1, 40)
```
